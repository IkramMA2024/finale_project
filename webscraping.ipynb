```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL de la page
url = "https://www.cdc.gov/mmwr/volumes/72/wr/mm7224a1.htm#:~:text=The%20age%2Dstandardized%20prevalence%20of%20depression%20was%20higher%20among%20women,(14.6%25)%2C%20and%20non%2D"

# Envoyer une requête GET à la page
response = requests.get(url)

# Vérifier que la requête a réussi
if response.status_code == 200:
    # Parser le contenu HTML
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Trouver tous les tableaux sur la page
    tables = soup.find_all('table')
    
    # Identifier le bon tableau (dans ce cas, le deuxième tableau)
    table = tables[1] if len(tables) > 1 else None

    # Vérifier si le tableau a été trouvé
    if table:
        # Extraire les en-têtes du tableau
        headers = [header.text.strip() for header in table.find_all('th')]

        # Extraire les lignes du tableau
        rows = table.find_all('tr')
        data = []
        for row in rows[1:]:
            cells = row.find_all('td')
            data.append([cell.text.strip() for cell in cells])

        # Créer un DataFrame pandas
        df = pd.DataFrame(data, columns=headers)

        # Sélectionner uniquement les colonnes d'intérêt
        df = df[['Jurisdiction', 'Sample size', 'Prevalence†\n% (95% CI)']]
        
        # Renommer les colonnes pour plus de clarté
        df.columns = ['Jurisdiction', 'Sample size', 'Proportion de dépression']
        
        # Afficher le DataFrame
        print(df)
    else:
        print("Tableau non trouvé.")
else:
    print("La requête a échoué avec le statut:", response.status_code)
```
